# The analysis of radiation pressure coefficient $\beta$ [1],[2]

Radiation pressure coefficient $\beta$ is the ratio of radiation and gravitational forces felt by an atom. We can also define $\beta$ as the ratio of the outward force of radiation pressure compared to the inner force of gravity. The ratio depends on the brightness of the central star at the wavelengths of the absorption lines of the species. [1] So we can define the ratio as:

$\begin{align}
\beta \equiv \frac{F_{rad}}{F_{grav}}
\end{align}$,

where the $F_{grav}$ is defined as

$\begin{align}
F_{grav} = \frac{G M m}{r^2}
\end{align}$

and $F_{rad}$ as:

$\begin{align}
F_{rad} = \frac{1}{c} \sum_{j < k} F_{\lambda} \sigma_{jk} \xrightarrow{\text{if distance-independent}} F_{rad} =\frac{1}{8 \pi c^2} \sum_{j < k} \frac{g_k}{g_j} A_{kj} \lambda_{jk}^4 F_{\lambda}.
\end{align}$

Both $F_{grav}$ and $F_{rad}$ are inverse square laws so that their ratio is independent of the distance to the star. So the equation for the ratio $\beta$ is:

$\begin{align}
\beta = \frac{1}{8 \pi c^2} \frac{r^2}{G M m} \sum_{j < k} \frac{g_k}{g_j} A_{kj} \lambda_{jk}^4 \left( F_{\lambda} \frac{R^2 4 \pi}{r^2 4 \pi} \cdot 4 \pi \right) \
\end{align}, $


The parameters from the equations are defined as followed:    
$\sigma_{jk}$ - cross-section for the transition between atomic levels $j$ and $k$ integrated over wavelength [$ cm^2$],    
$c$ - the speed of light [$cm/s$],    
$r$ - distance to the star at a given point in the orbit [$cm$],  
$R$ - radius of the chosen star [$cm$],     
$G$ - gravitational constant [$ cm^{3} \ g^{-1} \  s^{-2} $],        
$M$ - stellar mass [$g$],            
$m$ - mass of the atomic species under consideration [$g$],          
$k$ - higher atomic level of an atom [$dimensionless$],        
$j$ - lower atomic level of an atom [$dimensionless$],      
$g_k$ - statistical weight of level $k$ [$dimensionless$],          
$g_j$ - statistical weight of level $j$ [$dimensionless$],          
$A_{kj}$ - Einstein coefficient for spontaneous emission from level $k$ to $j$ [$s^{-1}$],          
$\lambda_{jk}$ -  the wavelength of the corresponding transition [$nm$],      
$F_{\lambda}$ - stellar flux density per unit wavelength evaluated at the line center [$erg \ s^{-1} \ cm^{-2} \ nm^{-1}$] $\rightarrow$ [$g \ s^{-3} \ nm^{-1} $].           

Unit erg is equal to $ \rightarrow [erg] = \left[ \frac{g \cdot cm^2}{s^2} \right]$

## Beta ratio uncertainty [2]

The uncertainties in the $\beta$ ratio were calculated using the following expression:

$\begin{align}
    \left( \frac{\delta \beta}{\beta} \right)^2 = 
\left( \frac{\delta \Phi}{\Phi} \right)^2_{\mathrm{cal}} + 
\frac{1}{\beta^2} \sum_i 
\left( \frac{g_{0,i}}{\sum g_0} \right)^2 \beta_i^2 
\left( \frac{\delta A_{i0}}{A_{i0}} \right)^2,
\end{align}$

where $\delta \beta$ is the uncertainty of the beta ratio. Here, $\beta$ represents the calculated ratio itself, $\delta \Phi / \Phi$ is the fractional uncertainty in the overall flux calibration, $g_0 / \sum g_0$ is the weighting factor accounting for the contribution of multiple ground states, and $A_{i0}$ is the Einstein coefficient corresponding to the $i$-th transition.

The uncertainty values for the $\beta$ ratio range from $4$% to $7$%. To ensure a conservative estimate. The uncertainties in the Einstein coefficients spanned from as low as $\leq 0.03$% for the strongest transitions to up to $100$% for the weakest ones. The wavelength uncertainties were within the range $4$% to $7$%, while the uncertainties in the weighted averages ranged from $0.3$% to $6$%.

The uncertainties of $\delta A_{i0}$ are taken from NIST database, where we corresponded value to given letters as followed:

## Velocity boost before ionization (for neutral atoms) [2]

The number of ionizing stellar photons hitting an atom per unit time is the stellar ionization rate:

$\begin{align}
\Gamma_* = \int_0^{\inf} \frac{F_{\lambda}}{hc/\lambda} (\sigma_{ion} m) d\lambda\ ,
\end{align} $

where           
$F_{\lambda}$ - stellar flux density per unit wavelength evaluated at the line center [$erg \ s^{-1} \ cm^{-2} \ nm^{-1}$],                   
$h$ - Planck constant [$cm^2 g/s$],      
$c$ - speed of light [$nm/s$],   
$m$ - mass of the atom [$g$],                        
$\sigma_{ion}$ - ionization cross-section [$cm^2/g$].    


The velocity boost before ionization is then

$\begin{align}
v_{ion} = \beta \frac{GM}{r^2} \frac{1}{\Gamma_*} \ ,
\end{align} $

where          
$G$ - gravitational constant [$ cm^{3} \ g^{-1} \  s^{-2} $],        
$M$ - stellar mass [$g$],                               
$\Gamma_*$ - stellar ionization rate [$1/s$].  

## Weakening effect on the Flux by the gas cloud between star and atom/ion [3]

Using Beer-Lambert Law:

$\begin{align}
F_{\lambda,gc} = F_{\lambda} \times e^{-\tau_{\lambda}}
\end{align}$

where optical depth at wavelength $\tau_{\lambda}$ is

$\begin{align}
\tau_{\lambda} = N \sigma_{\lambda}
\end{align}$

For bound-bound absorption (atomic, molecular lines), the cross section at specific wavelenghts is given as

$\begin{align}
\sigma_{\lambda} = \frac{e^2 \lambda_{tr}^2}{m_e c^2} f_{jk} \Phi(\lambda)
\end{align}$

where

$\begin{align}
\Phi(\lambda) = \frac{(\Gamma \lambda^2_{tr})/(4 \pi c)}{(\lambda - \lambda_{tr})^2 + (\Gamma \lambda^2_{tr})/(4 \pi c)}
\end{align}$

is Lorentzian line profile and damping constant $\Gamma$ is defined as

$\begin{align}
\Gamma = \sum_k A_{kj}
\end{align}$


The parameters from the equations are defined as followed:    
$F_{\lambda,gc}$ - stellar flux density per unit wavelength with added weakening effect evaluated at the line center [$erg \ s^{-1} \ cm^{-2} \ nm^{-1}$] $\rightarrow$ [$g \ s^{-3} \ nm^{-1} $],       
$F_{\lambda}$ - stellar flux density per unit wavelength evaluated at the line center [$erg \ s^{-1} \ cm^{-2} \ nm^{-1}$] $\rightarrow$ [$g \ s^{-3} \ nm^{-1} $],        
$\tau_{\lambda}$ - optical depth of specific atom/ion at certain wavelength [$unitless$],     
$N$ - column density of specific atom/ion [$cm^{-2}$],      
$\sigma_{\lambda}$ - cross-section of specific atom/ion at certain wavelength [$1/cm^{-2}$],     
$e$ - electron charge [$C$],   
$\lambda_{tr}$ - wavelength of the transition [$cm$],           
$m_e$ - mass of an electron [$g$],     
$c$ - the speed of light [$cm/s$],      
$f_{jk}$ - oscillator strength of the transition [unitless],     
$\Phi(\lambda)$ - line profile function [unitless],       
$\Gamma$ - damping constant [$s^{-1}$],            
$A_{kj}$ - Einstein coefficient of transition [$s^{-1}$],      
$\lambda$ - wavelength region of the flux [$cm$].   
         
Gas species under consideration are from [4],[5].

## Estimating the stellar wind strength [6]

Mass-loss rate of B-type stars:

$\begin{align}
\log \left(\frac{\dot{M}}{1 M_\odot\ \text{ year }^{-1}}\right) = -22.7 + 8.96 \left(\frac{T_{eff}}{10^4 \text{ K}}\right) - 1.42 \left(\frac{T_{eff}}{10^4 \text{ K}}\right)^2.
\end{align}$

Artificial velocity law:

$\begin{align}
v(r) = 10^{-3} \sqrt{\frac{5}{3} \frac{k T_{eff}}{m_H}} + 2 \cdot 10^8 \text{ cm/s } \frac{r-R_{star}}{R_{star}} \ ,
\end{align} $

where $T_{eff}$ is stellar effective temperature, $R_{star}$ stellar radius and $m_H$ is the mass of the hydrogen atom.

### Calculations

```python
###################CALCULATING THE SUM PART OF BETA RATIO, EQ (4)##########################
def first_equation(m):
    equation = []
    for el_m in mass:
        equation.append((1/(8*np.pi*c_cm**2))*((d_au**2)/(G_cm*sun_mass*m*el_m)))   # [s^4 cm^-3 g^-1]
    return equation

###########################################################################################

###################CALCULATING THE BETA RATIO, EQ (4)######################################
def beta_func(nm, Flambda, eq):
    beta_values = []
    uncertainties = []

    for y in range(len(file_name)):
        func = find_files(file_name[y])
        gk = func[0]
        g0 = func[1]
        A = func[2]          # [1/s]
        Aunc = func[3]
        L = func[4]          # [nm]
        el = []
        for x in range(len(L)):  # interpolate flux at transition wavelength
            corr = np.interp(L[x], nm, Flambda)
            el.append(gk[x] / g0[x] * A[x] * (L[x])**4 * corr * 1e-21)

        beta = eq[y] * sum(el)
        beta_values.append(beta)

        # --- Uncertainty estimation ---
        delta_phi_over_phi = 0.05  # 5% flux uncertainty
        omega_sum = np.sum(g0)

        sum_term = np.sum([(g0_i / omega_sum)**2 * (dAij)**2
                           for g0_i, dAij in zip(g0, Aunc)])
        
        rel_uncertainty_squared = delta_phi_over_phi**2 + sum_term
        uncertainties.append(beta*np.sqrt(rel_uncertainty_squared))

    return beta_values, uncertainties

###########################################################################################

###################GETTING ONLY THE NEUTRAL ELEMENT BETA RATIO######################################
def getting_neutral_beta(beta_values):
    neut_beta = [beta_values[0],beta_values[1],beta_values[3],beta_values[6],beta_values[9],beta_values[12],beta_values[15],
              beta_values[18],beta_values[21],beta_values[23],beta_values[26],beta_values[29],beta_values[32],beta_values[35],
              beta_values[38],beta_values[41],beta_values[44],beta_values[47],beta_values[50],beta_values[53],beta_values[56],
              beta_values[59],beta_values[62],beta_values[64],beta_values[66],beta_values[68],beta_values[71],beta_values[74]]
    return neut_beta

###########################################################################################

###################CALCULATING THE IONIZATION RATE, EQ (6)#################################
def ionization_rate(nm, Flambda):
    ion_rate = []
    for y in range(len(neut_atom)):
        # Get ionisation cross-section vs wavelength for this species
        cross_wavelengths, cross_sections = find(neut_atom[y], m_neutral[y])  # [nm], [cm²]
        # Define integrand: F_lambda(λ) * λ * σ_eff(λ)
        def integrand(lambda_nm):
            F_interp = np.interp(lambda_nm, nm, Flambda, left=0, right=0)        # [erg/s/cm²/nm]
            sigma_interp = np.interp(lambda_nm, cross_wavelengths, cross_sections, left=0, right=0)  # [cm²]
            return F_interp * lambda_nm * sigma_interp  # [erg·nm²/s/cm²]
        # Integrate over wavelengths (same units: nm)
        int_result, _ = quad(integrand, nm[0], nm[-1])

        # Final ionization rate: divide by hc (in erg·nm)
        Gamma = int_result / (h * c_nm)  # [1/s]
        ion_rate.append(Gamma)

    return ion_rate

###########################################################################################

###################CALCULATING THE VELOCITY BOOST, EQ (7)##################################
def velocity(neutral_beta, ion_rate, mass):
    v_ion = []
    for i in range(len(neutral_beta)):
            answer = neutral_beta[i] * ((G_cm*mass*sun_mass)/(d_au**2)) * (1/ion_rate[i])
            v_ion.append(answer*1e-5) # unit conversion [cm/s] --> [km/s]
    return v_ion

###########################################################################################
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ensure arrays once
s_nm = np.asarray(s_nm)

# --- YOUR TRANSITIONS (edit these) ---
TRANSITIONS = {
    "β Pic":       111,  
    "49 Ceti":     96,          
    "HD 209458":   170, 
    "KELT-9":      91,  
    "Kepler-1520": 350,          
}

# Optional: SciPy interpolation if available; otherwise fall back to numpy
try:
    from scipy.interpolate import interp1d as _interp1d
    def _interp(x, xp, fp):
        f = _interp1d(xp, fp, kind="linear", bounds_error=False, fill_value=np.nan, assume_sorted=False)
        return f(x)
except Exception:
    def _interp(x, xp, fp):
        xp = np.asarray(xp); fp = np.asarray(fp)
        order = np.argsort(xp)
        xp = xp[order]; fp = fp[order]
        return np.interp(x, xp, fp, left=np.nan, right=np.nan)

def _blend(y_short, y_long, x, x0, width):
    """Blend y_long (short-λ) into y_short (long-λ) around x0 with width."""
    w0 = x0 - width/2
    w1 = x0 + width/2
    ramp = (x - w0) / (w1 - w0)
    w = np.where(x <= w0, 0.0, np.where(x >= w1, 1.0, ramp))
    y_long_use  = np.where(np.isnan(y_long), 0.0, y_long)
    y_short_use = np.where(np.isnan(y_short), 0.0, y_short)
    mask_long   = ~np.isnan(y_long)
    mask_short  = ~np.isnan(y_short)
    y = (1 - w) * y_long_use + w * y_short_use
    only_long  = mask_long & ~mask_short
    only_short = ~mask_long & mask_short
    y[only_long]  = y_long_use[only_long]
    y[only_short] = y_short_use[only_short]
    y[~(mask_long | mask_short)] = np.nan
    return y

def _resolve_transition(cfg, wl_min, wl_max, x0_auto, width_auto):
    """Return (mode, x0, width, w0, w1). mode in {'hard','blend'}."""
    if cfg is None:
        # automatic (use your previous detection)
        x0 = float(np.clip(x0_auto, wl_min, wl_max))
        width = float(max(width_auto, 1e-6))
        return "blend", x0, width, x0 - width/2, x0 + width/2

    # number -> hard cut
    if isinstance(cfg, (int, float, np.floating)):
        x0 = float(np.clip(cfg, wl_min, wl_max))
        return "hard", x0, 0.0, x0, x0

    # pair/list -> blend window [start, end]
    if isinstance(cfg, (list, tuple)) and len(cfg) == 2:
        w0 = float(min(cfg))
        w1 = float(max(cfg))
        w0 = float(np.clip(w0, wl_min, wl_max))
        w1 = float(np.clip(w1, wl_min, wl_max))
        width = max(w1 - w0, 0.0)
        x0 = (w0 + w1) / 2.0
        if width <= 0.0:
            return "hard", x0, 0.0, x0, x0
        return "blend", x0, width, w0, w1

    # dict with centre/width or cut
    if isinstance(cfg, dict):
        if "cut" in cfg:
            x0 = float(np.clip(cfg["cut"], wl_min, wl_max))
            return "hard", x0, 0.0, x0, x0
        if "centre" in cfg:
            x0 = float(np.clip(cfg["centre"], wl_min, wl_max))
            width = float(cfg.get("width", width_auto))
            width = float(max(width, 1e-6))
            w0 = x0 - width/2
            w1 = x0 + width/2
            w0 = float(np.clip(w0, wl_min, wl_max))
            w1 = float(np.clip(w1, wl_min, wl_max))
            x0 = (w0 + w1) / 2.0
            width = max(w1 - w0, 1e-6)
            return "blend", x0, width, w0, w1

    # Fallback to auto if format unrecognised
    x0 = float(np.clip(x0_auto, wl_min, wl_max))
    width = float(max(width_auto, 1e-6))
    return "blend", x0, width, x0 - width/2, x0 + width/2

# ---- Main loop: interpolate, combine with chosen transition, save, and plot zoom ----
GRID_N = 80000
BLEND_FRAC = 0.01  # default blend width (fraction of full span) if you don't specify a width

combined_spec = {}
combined_zoom = {}

for (nm_t, f_t, scale, lab) in zip(nm_chosen, Flambda_chosen, scales, labels):
    nm_t = np.asarray(nm_t)
    f_t  = np.asarray(f_t)

    # Scale Sun for this target
    sun_scaled = sun_flux * scale

    # Common wavelength grid
    wl_min = float(np.nanmin([nm_t.min(), s_nm.min()]))
    wl_max = float(np.nanmax([nm_t.max(), s_nm.max()]))
    nm_common = np.linspace(wl_min, wl_max, GRID_N)

    # Interpolate
    y_atlas = _interp(nm_common, nm_t,  f_t)
    y_sun   = _interp(nm_common, s_nm, sun_scaled)

    # Automatic suggestion (as before)
    valid_atlas = ~np.isnan(y_atlas)
    if np.any(valid_atlas):
        idx_auto = int(np.argmax(valid_atlas))
        x0_auto = nm_common[idx_auto]
    else:
        x0_auto = 0.5 * (wl_min + wl_max)

    width_auto = BLEND_FRAC * (wl_max - wl_min)

    # --- Apply YOUR chosen transition for this target ---
    cfg = TRANSITIONS.get(lab, None)
    mode, x0, width, w0, w1 = _resolve_transition(cfg, wl_min, wl_max, x0_auto, width_auto)

    # Combine
    if mode == "hard" or width <= 0.0:
        # Hard switch at x0
        idx0 = int(np.searchsorted(nm_common, x0, side="left"))
        flux_combined = np.copy(y_sun)
        if idx0 < len(nm_common):
            flux_combined[idx0:] = y_atlas[idx0:]
    else:
        # Smooth blend centred at x0 with given width
        flux_combined = _blend(y_short=y_atlas, y_long=y_sun, x=nm_common, x0=x0, width=width)

    # ---- SAVE for later use ----
    combined_spec[lab] = {
        "nm": nm_common.copy(),
        "F_combined": flux_combined.copy(),
        "F_atlas_interp": y_atlas.copy(),
        "F_sun_interp": y_sun.copy(),
        "transition_mode": mode,
        "transition_nm": float(x0),
        "transition_width": float(width),
        "transition_window": (float(w0), float(w1)),
    }

    # ---------------- Transition zoom plot ----------------
    # Zoom window: around the transition window (or ±5*width for hard cut)
    if mode == "blend":
        span_left  = max(20.0, 2.5 * width)
        span_right = max(20.0, 2.5 * width)
    else:
        span_left  = max(20.0, 0.05 * (wl_max - wl_min))
        span_right = max(20.0, 0.08 * (wl_max - wl_min))

    x_lo = max(wl_min, x0 - span_left)
    x_hi = min(wl_max, x0 + span_right)

    in_full = (nm_common >= x_lo) & (nm_common <= x_hi)
    in_t    = (nm_t      >= x_lo) & (nm_t      <= x_hi)
    in_s    = (s_nm      >= x_lo) & (s_nm      <= x_hi)

    plt.figure(figsize=(10, 6))
    plt.plot(nm_common[in_full], flux_combined[in_full], label="Combined", color="black")
    if np.any(in_t):
        plt.plot(nm_t[in_t], f_t[in_t], label="ATLAS9", linestyle="--")
    if np.any(in_s):
        plt.plot(s_nm[in_s], sun_scaled[in_s], label="Scaled Sun (EUV)", linestyle=":")
    # Visual markers for the transition you chose
    if mode == "hard":
        plt.plot([x0, x0], [1e-20, 1e10], "--k", label=f"Transition: {x0:.1f} nm")
        plt.axvspan(x_lo, x0, alpha=0.15, color="grey", label="Sun-used region")
    else:
        plt.axvspan(w0, w1, alpha=0.12, color="grey", label=f"Blend: {w0:.1f}–{w1:.1f} nm")
        plt.plot([x0, x0], [1e-20, 1e10], ":", color="k")

    plt.xlabel("Wavelength (nm)")
    plt.ylabel(r"Flux  ($F_\lambda$)")
    plt.title(f"{lab}: Transition region (log scale)")
    plt.yscale("log")
    plt.xlim(0,400)
    plt.grid(True, which="both", alpha=0.25)
    plt.legend(frameon=False)
    plt.show()

```

## Getting HST spectra

```python
# =========================
# Configuration
# =========================
DATA_DIR = "/path/to/folder"

# filename, nominal range in nm
SEGMENTS_META = [
    ("HST_Bpic_115-145nm_COSFUV.fits", 115.0, 145.0),   # COS/FUV, x1dsum
    ("HST_BPic_162-315nm_STISNUV.fits",    162.0, 315.0),   # STIS/NUV, x1d
    ("HST_BPic_ 302-561nm_STISCCD.fits",    302.0, 561.0),   # STIS/CCD, x1d
    ("HST_BPic_545-1014nm_STISCCD.fits",    545.0, 1014.0),  # STIS/CCD, x1d
]

EDGE_GUARD_NM = 0.5      # trim at each band edge
SN_MIN = 2.0             # S/N threshold (relaxes to 0 if needed)
REBIN_DN = 0.2           # nm; set None to disable
NORM_BAND = (300.0, 320.0)

# =========================
# Helpers
# =========================
def load_x1d_clean(path, sn_min=SN_MIN, relaxed=False):
    """
    Return wavelength [nm], flux [erg s^-1 cm^-2 nm^-1] with quality masking.
    If relaxed=True, only finite & positive are enforced (no DQ/SN filters).
    """
    with fits.open(path) as hdul:
        wave = flux = err = dq = dq_wgt = None
        bunit = (hdul[0].header.get("BUNIT","") or "").lower()
        for h in hdul:
            if isinstance(h, fits.BinTableHDU):
                names = {n.upper(): n for n in h.columns.names}
                if "WAVELENGTH" in names and any(k in names for k in ("FLUX","NET","SCI")):
                    wave = np.hstack(h.data[names["WAVELENGTH"]])
                    for k in ("FLUX","NET","SCI"):
                        if k in names:
                            flux = np.hstack(h.data[names[k]])
                            bunit = (h.columns[names[k]].unit or bunit or "").lower()
                            break
                    err = np.hstack(h.data[names["ERROR"]]) if "ERROR" in names else None
                    dq = np.hstack(h.data[names["DQ"]]) if "DQ" in names else None
                    dq_wgt = np.hstack(h.data[names["DQ_WGT"]]) if "DQ_WGT" in names else None
                    break

    if wave is None or flux is None:
        raise ValueError(f"{os.path.basename(path)}: no WAVELENGTH/FLUX found")

    m = np.isfinite(wave) & np.isfinite(flux) & (flux > 0)
    if not relaxed:
        if err is not None:
            m &= np.isfinite(err)
            if sn_min and sn_min > 0:
                denom = np.where(err > 0, err, np.inf)
                m &= (np.abs(flux) / denom) >= sn_min
        if dq is not None:
            m &= (dq == 0)
        if dq_wgt is not None:
            m &= (dq_wgt > 0.5)

    w = wave[m]; f = flux[m]

    # Units: Å→nm ; per-Å→per-nm
    bunit = bunit.replace("angstrom","a")
    if "/a" in bunit or " a^-1" in bunit or bunit.endswith(" a-1"):
        f = f * 10.0
    if w.size and np.nanmedian(w) > 1000.0:
        w = w / 10.0

    return w, f


def clip_band(w, f, wmin, wmax, guard=EDGE_GUARD_NM):
    gmin, gmax = wmin + guard, wmax - guard
    if gmax <= gmin:
        return np.array([]), np.array([])
    m = (w >= gmin) & (w <= gmax)
    return w[m], f[m]


def rebin_avg(w_nm, f_per_nm, dn_nm):
    if dn_nm is None or dn_nm <= 0 or w_nm.size < 2:
        return w_nm, f_per_nm
    wmin, wmax = np.nanmin(w_nm), np.nanmax(w_nm)
    if not np.isfinite(wmin) or not np.isfinite(wmax) or wmax <= wmin:
        return w_nm, f_per_nm
    edges = np.arange(wmin, wmax + dn_nm, dn_nm)
    if edges.size < 2:
        return w_nm, f_per_nm
    bins = np.digitize(w_nm, edges) - 1
    valid = (bins >= 0) & (bins < len(edges) - 1)
    if not np.any(valid):
        return w_nm, f_per_nm
    bins = bins[valid]; f = f_per_nm[valid]
    num = np.bincount(bins, weights=f, minlength=len(edges)-1)
    den = np.bincount(bins, minlength=len(edges)-1)
    with np.errstate(invalid="ignore", divide="ignore"):
        fbin = num / den
    wbin = 0.5*(edges[:-1] + edges[1:])
    m = den > 0
    return wbin[m], fbin[m]


def scale_to_overlap(base_w, base_f, w, f):
    """Scale f to match base in their overlap; robust to sparse overlaps."""
    if base_w.size < 5 or w.size < 5:
        return f, 1.0
    lo = max(np.min(base_w), np.min(w))
    hi = min(np.max(base_w), np.max(w))
    if hi <= lo:
        return f, 1.0
    grid = np.linspace(lo, hi, 500)
    bf = np.interp(grid, base_w, base_f, left=np.nan, right=np.nan)
    cf = np.interp(grid, w, f, left=np.nan, right=np.nan)
    m = np.isfinite(bf) & np.isfinite(cf) & (bf > 0) & (cf > 0)
    if np.count_nonzero(m) < 20:
        return f, 1.0
    scale = np.nanmedian(bf[m] / cf[m])
    if not np.isfinite(scale) or scale <= 0:
        scale = 1.0
    return f * scale, scale


def normalise_band(w, f, w1, w2):
    sel = (w >= w1) & (w <= w2)
    if np.any(sel):
        med = np.nanmedian(f[sel])
        if np.isfinite(med) and med > 0:
            return f / med
    return f

# =========================
# Load, clean, clip (with fallback), rebin
# =========================
segments = []
for fname, w1, w2 in SEGMENTS_META:
    path = os.path.join(DATA_DIR, fname)
    # strict
    w, f = load_x1d_clean(path, sn_min=SN_MIN, relaxed=False)
    w, f = clip_band(w, f, w1, w2)
    # fallback if empty: relax quality cuts
    if w.size < 5:
        w_rel, f_rel = load_x1d_clean(path, sn_min=0.0, relaxed=True)
        w_rel, f_rel = clip_band(w_rel, f_rel, w1, w2)
        if w_rel.size >= 5:
            print(f"[info] Relaxed cuts used for {fname} (kept {w_rel.size} pts).")
            w, f = w_rel, f_rel
        else:
            print(f"[warn] Skipping {fname}: no usable data in {w1}-{w2} nm.")
            continue
    # (optional) light rebin per segment
    w, f = rebin_avg(w, f, REBIN_DN)
    segments.append((fname, w, f))

if not segments:
    raise RuntimeError("No usable segments after cleaning/clipping. Try lowering SN_MIN or EDGE_GUARD_NM.")

# =========================
# Scale-stitch by overlap
# =========================
segments_sorted = sorted(segments, key=lambda t: np.nanmin(t[1]))
ref_name, ref_w, ref_f = segments_sorted[0]
stitched_w = ref_w.copy()
stitched_f = ref_f.copy()
scales = {ref_name: 1.0}

for name, w, f in segments_sorted[1:]:
    f_scaled, s = scale_to_overlap(stitched_w, stitched_f, w, f)
    scales[name] = s
    stitched_w = np.concatenate([stitched_w, w])
    stitched_f = np.concatenate([stitched_f, f_scaled])
    order = np.argsort(stitched_w)
    stitched_w, stitched_f = stitched_w[order], stitched_f[order]

# Final gentle rebin for smoothness
W_HST, F_HST = rebin_avg(stitched_w, stitched_f, REBIN_DN)

# =========================
# Save absolute stitched spectrum
# =========================
df_hst = pd.DataFrame({"wavelength_nm": W_HST, "flux_erg_s_cm2_nm": F_HST})
#df_hst.to_csv("beta_pic_hst_stitched_clean.csv", index=False)
print("Saved beta_pic_hst_stitched_clean.csv")
print("Applied segment scales:")
for k,v in scales.items():
    print(f"  {k}: x{v:.4g}")

# =========================
# Optional ATLAS9 overlay (expects nmbpic, Flambdabpic in nm & per-nm)
# If model is per-Å, convert externally: Flambdabpic *= 10
# =========================
have_model = "nmbpic" in globals() and "Flambdabpic" in globals()
if have_model:
    w_mod = np.asarray(nmbpic, dtype=float)
    f_mod = np.asarray(Flambdabpic, dtype=float)
    m = np.isfinite(w_mod) & np.isfinite(f_mod)
    o = np.argsort(w_mod[m])
    w_mod, f_mod = w_mod[m][o], f_mod[m][o]

# =========================
# Plot (normalised for shape)
# =========================
plt.figure(figsize=(10,5))
plt.plot(W_HST, normalise_band(W_HST, F_HST, *NORM_BAND),
         label="HST (cleaned, clipped, scaled, rebinned)")
if have_model:
    plt.plot(w_mod, normalise_band(w_mod, f_mod, *NORM_BAND),
             label="ATLAS9 (normalised)")
plt.xlim(110, 650)
plt.xlabel("Wavelength (nm)")
plt.ylabel(f"Relative flux (normalised in {NORM_BAND[0]:.0f}–{NORM_BAND[1]:.0f} nm)")
plt.title("β Pictoris — COS + STIS stitched spectrum")
plt.legend()
plt.tight_layout()
plt.show()
```
